```python
from sklearn import datasets, mixture
from IPython.display import display
import pandas as pd
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
X = iris['data']
y = iris['target']
```

```python
X = pd.DataFrame(X)
print(X.shape)
display(X.head())
```

```python
gmm = mixture.GaussianMixture(n_components = 3)
```

```python
gmm.fit(X)
gmm.predict(X)
```

```python
print( gmm.aic(X) )  #The lower the better.
print( gmm.bic(X) ) 
```

```python
gmm.score(X)  
```

```python
gmm.predict_proba(X)
```

```python
gmm.sample(10) # sample from gmm
```
